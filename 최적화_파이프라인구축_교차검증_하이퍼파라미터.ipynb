## 최적화
- 목표 : 정확도를 96~98% 까지 올리시오!!! 
- 방법 
  - 데이터 증량(현재는 불가)
  - 품질향상을 고려(데이터 자체에 대해서 통찰)
    - 정규화(수치들간의 편차를 줄이기 위해서 특정구간으로 조정이 필요할까?)
    - 활성화 함수(sigmoid, softmax, tanh, Relu, Relu의 응용함수 등 ...) 등을 통해서 조정
    - 전처리기를 이용하여 처리 or 직접 구현하여서 처리

```python
X_train.shape
X_train[0]
# 데이터를 살펴보니 구간이 제법 넓다. 이 부분은 시각화로 그리거나, 
# df로 구성해서 통계요약으로 확인할수도 있다. -> 전처리기를 이용하여 구간을 좁혀보자
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# 전처리기를 통해서 데이터를 학습시키면 -> 정규화가 되어서 데이터가 나온다(전처리되서)
# 재료를 넣어서 최소최대스케일러의 전처리 세팅값을 준비한다. 
scaler = MinMaxScaler().fit(X_train)
```

```python
# 정규화 처리를 한다.(전처리 수행)
X_train_scaled = scaler.transform(X_train)
X_train_scaled[0]
# 데이터는 0~1사이로 정규화 되었다. (0: 최소, 1: 최대)
```

```python
# 알고리즘 새로 생성, 학습, 예측, 평가
clf2 = SVC()
clf2.fit(X_train_scaled, y_train)
y_pred2 = clf2.predict(X_test)
metrics.accuracy_score(y_test, y_pred2)
```

```python
# 예측 및 평가를 한번에 수행
clf2.score(X_test,y_test)
# 정확도가 더 떨어졌다? why? 훈련용 데이터를 전처리 했다. 그런데, 왜 테스트용 데이터는
# 전처리 안하지? => 데이터의 범위가 값의 범주가 변경되었다. 그런데 테스트 데이터는
# 그대로 사용한다? => 훈련용 데이터가 변화되었다면 테스트용도 같이 변화되어야한다. 
```

```python
# 테스트용 데이터도 전처리하여서 예측 수행한다.
# scaler.trasform(X_test) => 훈련용 데이터를 전처리하기 위해 준비된(최소,최대기준)
# 스케일러로 테스트용 데이터를 전처리한다. -> 최소 혹은 최대값이 넘어설 수도 있다.!!!
clf2.score(scaler.transform(X_test),y_test)
# 정확도가 우리 목표치 도달했다!!
# 98%가 목표치라면, 데이터 증설(X), 전처리(X), 알고리즘을 이제 살펴야한다.
# 이 알고리즘이 최적인가? 그 알고리즘 내에 하이퍼파라미터가 최적인가?
```

## 하이퍼 파라피터 튜닝

- 알고리즘이 어느 정도 선정되었다. 모형이 어느정도 준비가 되었다면, 알고리즘 자체를 최적화 하는 단계로 진입
- 매개변수 최적화(하이퍼파라미터 튜닝)

- 도구
  - validation_curve() : 단일 하이퍼파라미터 튜닝
  - **GridSearchCV : n개의 알고리즘의 하이퍼파라미터 튜닝(교차검증 도구 사용)**
  - ParameterGrid : 상동 

```python
from sklearn.model_selection import GridSearchCV
```

```python
# !pip install IPython 
```

```python
from IPython.display import Image
Image('res/하이퍼파라미터.png')
```

```python
# 구성 : 파라미터들을 딕셔너리로 나열 
# 기본값 중심으로 왼쪽은 기본값/배수, 오른쪽은 기본값*배수(예시)
# 이런 구성은 각자 다를 수 있다.
# 파라미터의 의미를 지금은 고려하지 않고, 방법에 집중
param_grid = {
    'C' : [ 0.001, 0.01, 0.1, 1, 10, 100, 1000 ],
    'gamma' : [0.001, 0.01, 0.1, 1, 10, 100, 1000] 
}
```

```python
# 교차 검증 생성 
# 위의 파라미터의 조합은 총 7x7 의 조합으로 알고리즘이 테스트 되고 
# 각각의 테스트 시, 데이터는 5fold로 분할되어서 5번의 훈련 및 테스트를 검증데이터가 바뀌면서 진행

# None, to use the default 5-fold cross validation
# fold => 내부적으로 훈련데이터는 5세트로 나눈다.
# 그 중 1개를 검증세트로 분류한다.
# 4개 세트에 대해서 훈련, 1개 세트에 대해서 테스트(평가, 예측) -> 예측결과 발생 (모델1, 예측결과1)
# 이런 과정을 검증포드를 변경하면서 계속 수행
# 결과 -> 예측결과 1, 예측결과2, ... , 예측결과5 => 평균 => 최종 예측 결과

# 교차검증 도구 생성
grid = GridSearchCV( SVC(), param_grid, cv=5)
```

```python
Image('res/머신러닝_학습시_검증폴드지정.png')
```

```python
# 훈련 
grid.fit(X_train_scaled,y_train )
```

```python
# 최적 파라미터 조합은? 
grid.best_params_
```

```python
# 최적 점수?
grid.best_score_
```

```python
# 테스트 데이터 (한번도 접하지 않은 데이터)를 가지고 평가
grid.score(scaler.transform(X_test),y_test)
```

- 한번도 접하지 못한 데이터를 가지고 예측해보니, 최고 점수 대비(98%) 정확도가 97% 정도로 떨어졌다. why???
- 왜 교차검증시 점수가 더 상승했지?
  - 이미 스케일된 데이터를 가지고 교차 검증을 수행했기 때문에, 검증데이터가 이미 노출 되었다 라는 의미!

## 파이프라인 구축

- 검증 폴드 문제점 해결( 검증용 데이터가 전처리시 이미 경험해 버리는 문제(이미 접해본 데이터) )
- 여러개의 알고리즘들을 비교 
- 하이퍼파라미터 튜닝도 같이 진행 가능 

```python
from sklearn.pipeline import make_pipeline, Pipeline
```

```python
# 생성방법 - 클래스
# 클래스는 키값들을 직접 지정, 함수형은 키값을 자동 제공 
# __init__ : 생섣자 
# verbose : 로그 관련
# 키 : 해당 재료들의 별칭, 제시하지 않으면 자동으로 값들을 소문자로 변경해서 만든다. 
# 리스트는 순서!!
pipe1 = Pipeline( [('scaler', MinMaxScaler()),('clf',SVC() )] )   
pipe1
```

```python
# 생성방법 - 함수
# 개별 estimators들의 별칭을 각각 이름을 소문자로 변경해서 표시 (클래스와 다른점)
pipe2 = make_pipeline(MinMaxScaler(), SVC())
pipe2 
```

```python
# 파이프라인 구축
# 1. 기본 생성자에서 가장 기본형을 생성하고
# 2. 교차검증 파라미터에서 비교대상들을 구성하고
# 3. 이들을 조합하여 교차검증 객체에 삽입
# 4. 훈련
# 5. 최적 파라미터 확인, 최적 점수 확인
# 6. 한번도 접하지 못한 데이터를 가지고 예측, 평가 
```

```python
# 0. 필요한 모듈 가져오기
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
```

```python
# 1. 기본 생성자에서 가장 기본형을 생성하고
pipe_std = Pipeline( [ ('prepocessing',MinMaxScaler), ('clf',SVC()) ] )
pipe_std
```

```python
# 2. 교차검증 파라미터에서 비교대상들을 구성하고
param_grid = [
    # 파라미터 내에 이름은 기준 파이프라인의 별칭을 따라간다.
    # 개별 알고리즘 내의 하이퍼파라미터는 별칭__파라미터명
    {
        'prepocessing' : [ StandardScaler(), MinMaxScaler() ], 
        'clf' : [SVC()],
        'clf__C' : [0.001, 0.01, 0.1 , 1, 10, 100, 1000],
        'clf__gamma' : [0.001, 0.01, 0.1 , 1, 10, 100, 1000]
    },
    {
        'prepocessing' : [None],
        'clf' : [RandomForestClassifier(n_estimators=100)],
        'clf__max_features' : [1,2,3]
    }
]
param_grid
# 차후 어느 정도 알고리즘들의 성향(특징)을 알게되면
# 교차검증에 필요한 대상들, 하이퍼파라미터를 더욱더 세밀하게 넣을 수 있다.
```

```python
# 3. 이들을 조합하여 교차검증 객체에 삽입
# 파이프라인에 기술된 기준대로, 교차검증을 수행하는데, 그 교차검증패턴은 param_grid에 
# 기술되어 있다. 그리고 훈련 데이터는 5fold로 구분되어 검증 절차를 차례대로 수행하게 된다.
grid = GridSearchCV( pipe_std, param_grid , cv=5) 
```

```python
# 4. 훈련
# X_train을 입력하는 이유는 파이프라인 따라서 교차 검증시 대상데이터만 전처리할거기 위해
# 이런 목적을 달성하려면 원본데이터가 들어와야한다.
grid.fit(X_train, y_train)
```

```python
# 5. 최적 파라미터 확인, 최적 점수 확인
grid.best_params_
```

```python
# 검증데이터로 검증 수행시 간섭없이(사전접촉없이) 잘 예측한 평균치
grid.best_score_
```

```python
# 6. 한번도 접하지 못한 데이터를 가지고 예측, 평가 
grid.score( X_test, y_test)
```
